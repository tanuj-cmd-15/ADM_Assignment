```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Example: df = pd.read_csv('your_dataset.csv')

# Calculate statistical measures for each numerical column
amzn_mean = df['AMZN'].mean()
amzn_median = df['AMZN'].median()
amzn_mode = df['AMZN'].mode()[0]

dpz_mean = df['DPZ'].mean()
dpz_median = df['DPZ'].median()
dpz_mode = df['DPZ'].mode()[0]

btc_mean = df['BTC'].mean()
btc_median = df['BTC'].median()
btc_mode = df['BTC'].mode()[0]

nflx_mean = df['NFLX'].mean()
nflx_median = df['NFLX'].median()
nflx_mode = df['NFLX'].mode()[0]

# Print the statistical measures
print("\nStatistical Measures:")
print("AMZN - Mean: {:.2f}, Median: {:.2f}, Mode: {:.2f}".format(amzn_mean, amzn_median, amzn_mode))
print("DPZ - Mean: {:.2f}, Median: {:.2f}, Mode: {:.2f}".format(dpz_mean, dpz_median, dpz_mode))
print("BTC - Mean: {:.2f}, Median: {:.2f}, Mode: {:.2f}".format(btc_mean, btc_median, btc_mode))
print("NFLX - Mean: {:.2f}, Median: {:.2f}, Mode: {:.2f}".format(nflx_mean, nflx_median, nflx_mode))

# Create boxplots for each numerical column
plt.figure(figsize=(16, 16))

plt.subplot(2, 2, 1)
sns.boxplot(y=df['AMZN'])
plt.title("Boxplot for AMZN")
plt.ylabel("AMZN")

plt.subplot(2, 2, 2)
sns.boxplot(y=df['DPZ'])
plt.title("Boxplot for DPZ")
plt.ylabel("DPZ")

plt.subplot(2, 2, 3)
sns.boxplot(y=df['BTC'])
plt.title("Boxplot for BTC")
plt.ylabel("BTC")

plt.subplot(2, 2, 4)
sns.boxplot(y=df['NFLX'])
plt.title("Boxplot for NFLX")
plt.ylabel("NFLX")

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import skew

amzn_skewness = skew(df['AMZN'])
dpz_skewness = skew(df['DPZ'])
btc_skewness = skew(df['BTC'])
nflx_skewness = skew(df['NFLX'])

# Function to determine skewness type
def skew_type(skewness):
    if skewness > 0:
        return "Positively Skewed"
    elif skewness < 0:
        return "Negatively Skewed"
    else:
        return "Symmetrical"

plt.figure(figsize=(16, 12))

plt.subplot(2, 2, 1)
sns.histplot(df['AMZN'], kde=True, bins=15, color='dodgerblue', label='AMZN Distribution', alpha=0.7)
plt.axvline(df['AMZN'].mean(), color='darkred', linestyle='--', label=f'Mean: {df["AMZN"].mean():.2f}', linewidth=1.5)
plt.axvline(df['AMZN'].median(), color='darkgreen', linestyle='-', label=f'Median: {df["AMZN"].median():.2f}', linewidth=1.5)
plt.axvline(df['AMZN'].mode()[0], color='darkblue', linestyle=':', label=f'Mode: {df["AMZN"].mode()[0]:.2f}', linewidth=1.5)
plt.title("AMZN Distribution with Mean, Median, and Mode", fontsize=14, fontweight='bold')
plt.xlabel("AMZN", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(fontsize=10)
plt.text(df['AMZN'].max() * 0.6, 8, f"Skewness: {amzn_skewness:.2f}\n({skew_type(amzn_skewness)})", 
         fontsize=10, color='purple', bbox=dict(facecolor='white', alpha=0.7))

plt.subplot(2, 2, 2)
sns.histplot(df['DPZ'], kde=True, bins=15, color='salmon', label='DPZ Distribution', alpha=0.7)
plt.axvline(df['DPZ'].mean(), color='darkred', linestyle='--', label=f'Mean: {df["DPZ"].mean():.2f}', linewidth=1.5)
plt.axvline(df['DPZ'].median(), color='darkgreen', linestyle='-', label=f'Median: {df["DPZ"].median():.2f}', linewidth=1.5)
plt.axvline(df['DPZ'].mode()[0], color='darkblue', linestyle=':', label=f'Mode: {df["DPZ"].mode()[0]:.2f}', linewidth=1.5)
plt.title("DPZ Distribution with Mean, Median, and Mode", fontsize=14, fontweight='bold')
plt.xlabel("DPZ", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(fontsize=10)
plt.text(df['DPZ'].max() * 0.6, 8, f"Skewness: {dpz_skewness:.2f}\n({skew_type(dpz_skewness)})", 
         fontsize=10, color='purple', bbox=dict(facecolor='white', alpha=0.7))

plt.subplot(2, 2, 3)
sns.histplot(df['BTC'], kde=True, bins=15, color='limegreen', label='BTC Distribution', alpha=0.7)
plt.axvline(df['BTC'].mean(), color='darkred', linestyle='--', label=f'Mean: {df["BTC"].mean():.2f}', linewidth=1.5)
plt.axvline(df['BTC'].median(), color='darkgreen', linestyle='-', label=f'Median: {df["BTC"].median():.2f}', linewidth=1.5)
plt.axvline(df['BTC'].mode()[0], color='darkblue', linestyle=':', label=f'Mode: {df["BTC"].mode()[0]:.2f}', linewidth=1.5)
plt.title("BTC Distribution with Mean, Median, and Mode", fontsize=14, fontweight='bold')
plt.xlabel("BTC", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(fontsize=10)
plt.text(df['BTC'].max() * 0.6, 8, f"Skewness: {btc_skewness:.2f}\n({skew_type(btc_skewness)})", 
         fontsize=10, color='purple', bbox=dict(facecolor='white', alpha=0.7))

# NFLX Distribution
plt.subplot(2, 2, 4)
sns.histplot(df['NFLX'], kde=True, bins=15, color='orange', label='NFLX Distribution', alpha=0.7)
plt.axvline(df['NFLX'].mean(), color='darkred', linestyle='--', label=f'Mean: {df["NFLX"].mean():.2f}', linewidth=1.5)
plt.axvline(df['NFLX'].median(), color='darkgreen', linestyle='-', label=f'Median: {df["NFLX"].median():.2f}', linewidth=1.5)
plt.axvline(df['NFLX'].mode()[0], color='darkblue', linestyle=':', label=f'Mode: {df["NFLX"].mode()[0]:.2f}', linewidth=1.5)
plt.title("NFLX Distribution with Mean, Median, and Mode", fontsize=14, fontweight='bold')
plt.xlabel("NFLX", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend(fontsize=10)
plt.text(df['NFLX'].max() * 0.6, 8, f"Skewness: {nflx_skewness:.2f}\n({skew_type(nflx_skewness)})", 
         fontsize=10, color='purple', bbox=dict(facecolor='white', alpha=0.7))

plt.tight_layout()
plt.show()
```
